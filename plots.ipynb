```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the predictions CSV file.
df_syn = pd.read_csv('./BayesianLinearRegression/synthetic_Sigmoidal_bayesian_predictions_tuned.csv')

# Assume the file has a column named "x" for the input feature.
plt.figure(figsize=(8, 6))
plt.plot(df_syn['features'], df_syn['true'], 'bo', label='True values')
plt.plot(df_syn['features'], df_syn['prediction'], 'r-', label='Predictions')
plt.fill_between(df_syn['features'], df_syn['lower95'], df_syn['upper95'], color='gray', alpha=0.3, 
                 label='95% Confidence Interval')
plt.xlabel('Features')
plt.ylabel('Target Value')
plt.title('Synthetic Data: Predictions with 95% Confidence Intervals')
plt.legend()
# plt.savefig('./results/synthetic_predictions_plot.png')
plt.show()
```

```python
# Load the weights CSV file.
df_weights = pd.read_csv('./BayesianLinearRegression/synthetic_Sigmoidal_bayesian_weights_tuned.csv')

plt.figure(figsize=(10, 6))
plt.bar(df_weights['Index'], df_weights['Weight'], color='skyblue')
plt.xlabel('Basis Function Index')
plt.ylabel('Posterior Weight Value')
plt.title('Wine Quality (Red): Learned Posterior Weights')
# plt.savefig('./results/wine_red_weights_plot.png')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Needed to enable 3D plots in matplotlib

# 1. Load your dataset (update the file path and column names to match your CSV)
df = pd.read_csv('./BayesianLinearRegression/forestfires_Polynomial_bayesian_predictions_tuned.csv')

# Example column names: 'TrueValue', 'PredValue', 'CI_Lower', 'CI_Upper'
true_vals = df['true'].values
pred_vals = df['prediction'].values
ci_lower = df['lower95'].values
ci_upper = df['upper95'].values

# 2. Create some x-axis for each data point, for example the row index
idx = np.arange(len(df))

# 3. Initialize 3D figure and axes
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# 4. Scatter-plot the (true vs. predicted) vs. index in 3D
#    Here we use:
#       x-axis    = true values
#       y-axis    = predicted values
#       z-axis    = an index (just so we can see each point in 3D)
scatter_plot = ax.scatter(true_vals, pred_vals, idx,
                          c='blue', alpha=0.7, label='Data Points')

# 5. Plot “error bar” lines for the confidence interval, 
#    drawn parallel to the y-axis for each point
for i in range(len(df)):
    ax.plot(
        [true_vals[i], true_vals[i]],       # x stays the same for the point
        [ci_lower[i], ci_upper[i]],         # y goes from lower to upper CI
        [idx[i], idx[i]],                   # z stays the same for the point
        color='gray', alpha=0.5
    )

# 6. Label axes
ax.set_xlabel('True Value')
ax.set_ylabel('Predicted Value')
ax.set_zlabel('Index (Row Number)')

# 7. Optional: set a title
ax.set_title('3D Plot: Prediction vs. True with 95% Confidence Intervals')

ax.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load cost history data
poly_cost = pd.read_csv('./results/synthetic_poly_cost_history.csv')
sigmoid_cost = pd.read_csv('./results/synthetic_sigmoid_cost_history.csv')
gaussian_cost = pd.read_csv('./results/synthetic_gaussian_cost_history.csv')

# Load prediction data
poly_predictions = pd.read_csv('./results/synthetic_poly_predictions.csv')
sigmoid_predictions = pd.read_csv('./results/synthetic_sigmoid_predictions.csv')
gaussian_predictions = pd.read_csv('./results/synthetic_gaussian_predictions.csv')
```

```python
# Plot MSE over iterations for each basis function
plt.figure(figsize=(12, 6))

# Polynomial Basis
plt.plot(poly_cost['Iteration'], poly_cost['Cost'], label='Polynomial Basis', color='blue')

# Sigmoidal Basis
plt.plot(sigmoid_cost['Iteration'], sigmoid_cost['Cost'], label='Sigmoidal Basis', color='green')

# Gaussian Basis
plt.plot(gaussian_cost['Iteration'], gaussian_cost['Cost'], label='Gaussian Basis', color='red')

# Add labels, title, and legend
plt.xlabel('Iteration')
plt.ylabel('Cost (MSE)')
plt.title('MSE Over Iterations for Logistic Regression')
plt.legend()
plt.grid()
plt.show()
```

```python
def plot_predictions(data, title, color):
    plt.scatter(data['x1'], data['x2'], c=data['true_label'], cmap='coolwarm', label='True Labels', alpha=0.6)
    plt.scatter(data['x1'], data['x2'], c=data['predicted_label'], cmap='coolwarm', marker='x', label='Predicted Labels', alpha=0.6)
    plt.title(title)
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.legend()
    plt.grid()
```

```python
plt.figure(figsize=(18, 6))

# Polynomial Basis
plt.subplot(1, 3, 1)
plot_predictions(poly_predictions, 'Polynomial Basis Predictions', 'blue')

# Sigmoidal Basis
plt.subplot(1, 3, 2)
plot_predictions(sigmoid_predictions, 'Sigmoidal Basis Predictions', 'green')

# Gaussian Basis
plt.subplot(1, 3, 3)
plot_predictions(gaussian_predictions, 'Gaussian Basis Predictions', 'red')

plt.tight_layout()
plt.show()
```
